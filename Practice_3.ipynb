<h3 align='center'>Сортування IP-адрес</h3>
IP-адреса – унікальний числовий ідентифікатор пристрою в комп'ютерній мережі, що працює за протоколом TCP/IP.

У 4-й версії IP-адреса являє собою 32-бітове число. Адреса записується у вигляді чотирьох десяткових чисел (октетів) зі значенням від 00 до 255, розділених точками, наприклад, 192.168.1.2.
Напишіть програму, яка зчитує IP-адреси та виводить їх у порядку зростання відповідно до десяткового подання.

<b>Формат вхідних даних</b><br>
На вхід до програми подається число (n≥1) – кількість IP-адрес. Потім n рядків із коректними IP-адресами.

<b>Формат вихідних даних</b><br>
Програма повинна вивести IP-адреси у порядку зростання відповідно до десяткового подання.

<b>Примітка 1.</b> Щоб перевести IP-адресу 192.168.1.2 до десяткового числа, ми використовуємо формулу:




<b>Примітка 2.</b> Використовуйте додатковий аргумент key.

Тестовые данные 🟢<br>
<b>Sample Input:</b><br>

128.199.44.24<br>

<h3 align='center'>Код програми</h3>

```python
import ipaddress
n = int(input())
spisok = []
for i in range(n):
    temp = input()
    spisok.append(temp)
sortedkey = sorted(spisok, key = ipaddress.IPv4Address)
print('\n\n')
print(*sortedkey, sep='\n')
```
