<h3 align='center'>Таблиця множення</h3>
На вхід до програми подаються два натуральні числа nn і mm — кількість рядків і стовпців у матриці. Створіть матрицю mult розміром n n×m та заповніть її таблицею множення за формулою mult[i][j] = i * j.

<b>Формат вхідних даних</b><br>
На вхід до програми на різних рядках подаються два числа n і m — кількість рядків і стовпців у матриці.

<b>Формат вихідних даних</b><br>
Програма повинна вивести таблицю множення, відводячи на виведення кожного числа рівно 3 символи (для цього використовуйте рядковий метод ljust()).

```python
# put your python code here
n = int(input())
m = int(input())
matrix = [[0]*m for _ in range(n)] 
for i in range(n):
    for j in range(m):
        matrix[i][j] = i * j
for i in range(n):        
    print(*matrix[i])
```

<h3>Максимум у таблиці</h3>
На вхід програмі подаються два натуральні числа nn і mm — кількість рядків і стовпців у матриці, потім nn рядків по mm цілих чисел у кожному, відокремлених символом пропуску.

Напишіть програму, яка знаходить індекси (рядок та стовпець) першого входження максимального елемента.

<b>Формат вхідних даних</b><br>
На вхід до програми на різних рядках подаються два числа nn і mm — кількість рядків і стовпців у матриці, потім самі елементи матриці рядково через пробіл.

<b>Формат вихідних даних</b><br>
Програма має вивести два числа: номер рядка та номер стовпця, у яких стоїть найбільший елемент таблиці. Якщо таких елементів кілька, то виводиться той, у якого менший номер рядка, а якщо номери рядків рівні, той, у якого менший номер стовпця.

<b>Примітка</b>. Вважайте, що нумерація рядків та стовпців починається з нуля.

```python
# put your python code here
n = int(input())
m = int(input())
maxn = -9999
matrix = [list(map(int, input().split())) for _ in range(n)]
for i in range(n):
    for j in range(m):
        if matrix[i][j] > maxn:
            maxn = matrix[i][j]
for i in range(n):
    if maxn in matrix[i]:
        print('')
        print(i, matrix[i].index(maxn))
        break
```

<h3>Обмін стовпців</h3>
Напишіть програму, яка змінює місцями стовпці у матриці.

<b>Формат вхідних даних</b><br>
На вхід програмі на різних рядках подаються два натуральні числа n і m - кількість рядків і стовпців у матриці, потім елементи матриці рядкове через пробіл, потім числа i і j - номери стовпців, що підлягають обміну.

<b>Формат вихідних даних</b><br>
Програма повинна вивести вказану таблицю із заміненими стовпцями.

```python
# put your python code here
n = int(input())
m = int(input())
matrix = [input().split() for _ in range(n)]
x = input().split()
i = int(x[0])
j = int(x[1])
for k in range(n):
    matrix[k][i], matrix[k][j] = matrix[k][j], matrix[k][i]
print('')
for i in range(n):  
    print(*matrix[i])
```

<h3 align='center'>Симетрична матриця</h3>
Напишіть програму, яка перевіряє симетричність квадратної матриці щодо головної діагоналі.

<b>Формат вхідних даних</b><br>
На вхід до програми подається натуральне число nn — кількість рядків і стовпців у матриці, потім елементи матриці рядково через пропуск.

<b>Формат вихідних даних</b><br>
Програма повинна вивести YES, якщо матриця симетрична щодо головної діагоналі, і слово NO інакше.

```python
# put your python code here
n = int(input())
matrix = [input().split() for _ in range(n)]
flag = True
for i in range(n):
    for j in range(n): 
        if matrix[i][j] == matrix[j][i]: continue
        else: flag = False
print('')
if flag: print('YES')
else: print('NO')
```

<h3 align='center'>Обмін діагоналей</h3>
Дано квадратну матрицю чисел. Напишіть програму, яка змінює місцями елементи, що стоять на головній та побічній діагоналі, при цьому кожен елемент повинен залишитися в тому ж стовпці (тобто в кожному стовпці потрібно поміняти місцями елемент на головній діагоналі та побічній діагоналі).

<b>Формат вхідних даних</b><br>
На вхід до програми подається натуральне число nn — кількість рядків і стовпців у матриці, потім елементи матриці рядково через пропуск.

<b>Формат вихідних даних</b><br>
Програма повинна вивести матрицю з елементами головної та побічної діагоналі, що змінилися своїми місцями.

```python
# put your python code here
n = int(input())
matrix = [input().split() for _ in range(n)]
for i in range(n):
    for j in range(n): 
        if i == j:
            matrix[i][j], matrix[n - i - 1][j] = matrix[n - i - 1][j], matrix[i][j]
print('')
for i in range(n):  print(*matrix[i]) 
```

<h3 align='center'>Поворот матриці</h3>
Напишіть програму, яка повертає квадратну матрицю чисел на 90<sup>o</sup>за годинниковою стрілкою.

<b>Формат вхідних даних</b><br>
На вхід до програми подається натуральне число nn — кількість рядків і стовпців у матриці, потім елементи матриці рядково через пропуск.

<b>Формат вихідних даних</b><br>
Програма повинна вивести результат на екран, числа мають бути розділені одним пропуском.

```python
# put your python code here
n = int(input())
matrix = [input().split() for _ in range(n)]
matrix.reverse()
for i in range(n):
    print('')
    for j in range(n):
        print(matrix[j][i], end=' ')
```

<h3 align='center'>Ходи коня</h3>
На шахівниці 8×8 стоїть кінь. Напишіть програму, яка відзначає положення коня на дошці та всі клітини, які б'є кінь. Клітку, де стоїть кінь, відзначте англійською літерою N, клітини, які б'є кінь, позначте символами *, решту клітин заповніть крапками.

<b>Формат вхідних даних</b><br>
На вхід програмі подаються координати коня на шахівниці в шахової нотації (тобто у вигляді e4, де спочатку записується номер стовпця (літера від a до h, зліва направо), потім номер рядка (цифра від 1 до 8, знизу вгору)).

<b>Формат вихідних даних</b><br>
Програма має вивести на екран зображення дошки, розділяючи елементи пробілами.

<b>Примітка</b>. Шахова дошка має вигляд:


```python
# put your python code here
xy = input()
x = '87654321'.index(xy[1])
y = 'abcdefgh'.index(xy[0])
matrix = [['.']*8 for _ in range(8)]
matrix[x][y] = 'N'
for i in range(8):
    for j in range(8):
        if ((x-j)*(y-i) == 2) or ((x-j)*(y-i) == -2):
            matrix[j][i] = '*'
for i in range(8):
    print(*matrix[i])
```

<h3 align='center'>Магічний квадрат 🌶️</h3>
Магічним квадратом порядку nn називається квадратна таблиця розміру n×n, складена з усіх чисел 1, 2, 3,... n<sup>2</sup> так, що суми по кожному стовпцю, кожному рядку та кожній із двох діагоналей рівні між собою. Напишіть програму, яка перевіряє, чи є задана квадратна матриця магічним квадратом.

<b>Формат вхідних даних</b><br>
На вхід до програми подається натуральне число nn — кількість рядків і стовпців у матриці, потім елементи матриці: nn рядків, по nn чисел у кожному, розділені пробілами.

<b>Формат вихідних даних</b><br>
Програма повинна вивести слово YES, якщо матриця є магічним квадратом, і слово NO інакше.

```python
# put your python code here
n = int(input())
mtx = [list(map(int, input().split())) for _ in range(n)]
lst = [row[i] for row in mtx for i in range(n)]

row = [sum(row) for row in mtx]
col = [sum([mtx[i][j] for i in range(n)]) for j in range(n)]
main = sum([mtx[i][i] for i in range(n)])
sub = sum([mtx[i][n-i-1] for i in range(n)])
print('')
print(('NO', 'YES')[all([i in lst for i in range(1, n**2 + 1)]) and row == col and main == sub and row[1] == main])
```
