<h3 align='center'> Метод Монте-Карло </h3>

Применим метод Монте-Карло і до задачі вичислення площі геометричної фігури на плоскості.

Помістимо фігуру в прямоугольник і будемо наугад кидати точки в цей прямоугольник. Будем виходити з того, що чим більша площа фігури, тим більше в неї будуть потрапляти точки. Таким чином, при великому числі nn точок, наугад вибраних всередині прямоугольника вмісту, доля точок, що знаходяться в даній фігурі kk, приближено до рівного розташування площі цієї фігури і площі прямоугольника.

<h3 align='center'> Завдння </h3>

Напишіть програму, яка за допомогою методу Монте-Карло обчислює площу фігури, що задається за допомогою системи нерівностей:
    

<h3 align='center'>Код програми</h3>

```python
import random

n = 10**6       # кількість випробувань
k = 0
s0 = 16
for _ in range(n):
    x = random.uniform(-2, 2)
    y = random.uniform(-2, 2)

    if x**3 + y**4 + 2 >= 0 and 3*x + y**2 <= 2:
        k += 1
print((k/n)*s0)
```
