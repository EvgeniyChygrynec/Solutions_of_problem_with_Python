<h3 align='center'>Пропущені коментарі 🌶️</h3>

При написанні власних функцій рекомендується в коментарі описувати призначення функції, її параметри та значення, що повертається. Часто програмісти відкладають написання таких коментарів наостанок, а потім взагалі забувають про них 😂.

На вхід до програми подається рядок тексту з ім'ям текстового файлу, в якому написаний код мовою Python. Напишіть програму, яка виводить на екран імена всіх функцій для яких відсутній пояснювальний коментар. Вважатимемо, що будь-який рядок, що починається зі слова def і пробілу, є початком визначення функції. Функція містить коментар, якщо перший символ попереднього рядка – #.

<h4>Формат вхідних даних</h4>
На вхід програмі подається рядок тексту, що містить ім'я існуючого текстового файлу з кодом Python.

<h4>Формат вихідних даних</h4>
Програма повинна вивести назви всіх функцій (не змінюючи порядку їхнього прямування у вихідному файлі), кожна на окремому рядку, для яких відсутній пояснювальний коментар. Якщо всі функції у файлі мають пояснювальний коментар, слід вивести: Best Programming Team.

<b>Примітка 2.</b> Гарантується, що файл має хоча б одну функцію при цьому вкладених функцій у файлі немає.

<h3 align='center'>Код програми</h3>

```python
with open(input(), encoding='utf-8') as inf:
    not_commented_funcs, preline = [], ''
    for line in inf:
        if not preline.startswith('#') and line.startswith('def '):
            not_commented_funcs.append(line[4:line.find('(')])
        preline = line
    print('\n'.join(not_commented_funcs) if not_commented_funcs else 'Best Programming Team')
```
