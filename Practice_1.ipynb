<h3 align='center'>Максимальний в області 2</h3>   
<p>Напишіть програму, яка виводить максимальний елемент у заштрихованій області квадратної матриці.</p>



<h4>Формат вхідних даних</h4>
На вхід до програми подається натуральне число n — кількість рядків і стовпців у матриці, потім елементи матриці.

<h4>Формат вихідних даних</h4>
Програма повинна вивести одне число - максимальний елемент у області квадратної матриці.

Примітка. Елементи побічної діагоналі також враховуються.

Тестовые данные 🟢 
<hr width=800px align='left'>
<h4>Sample Input 1:</h4>

3<br>
1 4 5<br>
6 7 8<br>
1 1 6

<h4>Sample Output 1:</h4>
8
<hr width=800px align='left'>

<h4>Sample Input 2:</h4>

4<br>
0 1 4 6<br>
0 0 2 5<br>
0 0 0 7<br>
0 0 0 0

<h4>Sample Output 2:</h4>

7



```python
list = [] 
print('Введіть кількість рядків і стовпців у матриці:')
n = int(input())
matrix = []
print('\nВведіть елементи матриці:')
for i in range(n):
    temp = [int(num) for num in input().split()]
    matrix.append(temp)
for i in range(len(matrix)):
    for j in range(len(matrix)):
        if i >= n - 1 -j:
            list.append(matrix[i][j])

print(f'\nМаксимальний елемент у області квадратної матриці: {max(list)}')
```
