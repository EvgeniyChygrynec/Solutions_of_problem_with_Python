<h3 align='center'>Метод Монте-Карло</h3>

Напишіть програму, яка за допомогою методу Монте-Карло визначає наближене значення числа π.

<b>Примітка 1.</b> Площа одиничного кола, тобто кола з радіусом, що дорівнює R = 1 дорівнює S = πR<sup>2</sup> = π.

<b>Примітка 2.</b> Рівняння одиничного кола має вигляд x<sup>2</sup>+y<sup>2</sup> = 1

<h3 align='center'>Код програми</h3>

```python
import random

n = 10**6       # кількість випробувань
k = 0
s0 = 4
for _ in range(n):
    x = random.uniform(-1, 1)
    y = random.uniform(-1, 1)
    
    if x**2 + y**2 <= 1:
        k += 1
        
print((k/n)*s0)
```
